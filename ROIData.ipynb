```python
import numpy as np
import matplotlib.pyplot as plt

# Time in months
months = np.arange(0, 60, 1)

# Parameters
n_sats = 12
cost_per_sat = 45e6  # $45 million
total_cost = n_sats * cost_per_sat

# Simulated data revenue over time (nonlinear growth)
monthly_revenue = 5e6 * (1 - np.exp(-months / 12)) + 1e6 * np.random.rand(len(months))
cumulative_revenue = np.cumsum(monthly_revenue)

# ROI calculation
roi = (cumulative_revenue - total_cost) / total_cost * 100

# Plotting
plt.figure(figsize=(12, 6))

# Revenue & Cost
plt.subplot(1, 2, 1)
plt.plot(months, cumulative_revenue / 1e6, label='Cumulative Revenue ($M)', color='green')
plt.axhline(y=total_cost / 1e6, color='red', linestyle='--', label='Total Investment Cost ($M)')
plt.title('Revenue vs Investment')
plt.xlabel('Time (months)')
plt.ylabel('Millions USD')
plt.legend()
plt.grid(True)

# ROI
plt.subplot(1, 2, 2)
plt.plot(months, roi, label='ROI (%)', color='blue')
plt.axhline(0, color='gray', linestyle='--')
plt.title('Return on Investment Over Time')
plt.xlabel('Time (months)')
plt.ylabel('ROI (%)')
plt.grid(True)

plt.suptitle('ROI Simulation for Satellite Data Antenna Array', fontsize=15)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
